```python
import os
from email.parser import Parser


#Link to directory of information and looking into Beck S's Australia folder
rootdir1 = r'/home/EDSE-suaniarh-workspace/Enron_dataset/maildir/beck-s/australia/4'

#Creating a helper function to eventually move data from Beck's Australia folder into a list
def email_analyse(inputfile, to_email_list, from_email_list, email_body):
    with open(inputfile, "r") as f:
        data = f.read()  # reads the file
 
    email = Parser().parsestr(data) 
 
    to_email_list.append(email['to'])
    from_email_list.append(email['from'])
 
    email_body.append(email.get_payload())
    
    
#Emptying lists to structure the data!!!
to_email_list = []
from_email_list = []
email_body = []

```

```python
'''

PART 1: Use Helper Function to move data from unstructured email form to 
structured form

'''

#Calling out the email_analyse function created above to move data from an unstructured to a structured format
print ("moving data to new structures")
for directory, subdirectory, filenames in  os.walk(rootdir1):
    for filename in filenames:
        #using the root directory leading to Beck's autralia folder
        email_analyse(os.path.join(directory, filename), to_email_list, from_email_list, email_body )    
```

```python
'''
PART 2: Moving the structured data into txt files
'''

#Moving the "to_email_list"
print ("saving to email list as text \n")
with open("to_email_list.txt", "w") as f:
    for to_email in to_email_list:
        if to_email: #Only if there is a to email then write. This avoids adding empty lines
            f.write(to_email)
            f.write("\n")

#Moving the from_email list
print ("saving from email list as text \n")
with open("from_email_list.txt", "w") as f:
    for from_email in from_email_list:
        if from_email: #Only if there is a from email then write. This avoids adding empty lines
            f.write(from_email)
            f.write("\n")        

#Moving email body
print ("saving email body as text \n")
with open("email_body.txt", "w") as f:
    for email_bod in email_body:
        if email_bod: #if there is an email body then write. This avoids adding empty lines
            f.write(email_bod)
            f.write("\n")  
```

```python
##Adding data to lists using the helper function, email_analyse1

import os
from collections import Counter
import time

from email.parser import Parser
rootdir2 = r'/home/EDSE-suaniarh-workspace/Enron_dataset/maildir/beck-s/australia' 

##Creating the following three lists to do analysis: (1) to_email_list, (2) from_email_list, and (3) email_body.

to_email_list = []
from_email_list = []
email_body = []


def email_analyse1(inputfile, to_email_list, from_email_list, email_body):
    with open(inputfile, "r") as f:
        data = f.read()

    email = Parser().parsestr(data)
      
    if email['to']:
        email_to = email['to']
        email_to = email_to.replace("\n", "")
        email_to = email_to.replace("\t", "")
        email_to = email_to.replace(" ", "")

        email_to = email_to.split(",")

        for email_to_1 in email_to:
            to_email_list.append(email_to_1)

    from_email_list.append(email['from'])

    email_body.append(email.get_payload())



start = time.time()

for directory, subdirectory, filenames in  os.walk(rootdir2):
    for filename in filenames:
        email_analyse1(os.path.join(directory, filename), to_email_list, from_email_list, email_body )

print("\nTo email adresses: \n")
print(Counter(to_email_list).most_common(10), "\n") #From Collections list no more than 10 To_Email items. 
#10 Names names came up: Sally, Heidi... etc.As you can see multiple emails were
#sent to individual email addresses.Most_common() method in this line of code organizes the 
#data according to the number of sent emails per individual in the descending order

print("The number of emails sent = ", len(to_email_list)) #This tells there were 9 emails sent altogether. 

print("\nFrom email adresses: \n")
print(Counter(from_email_list).most_common(10), "\n") #This does the same thing as in the code above,
#except that it returns all the from email addresses in the descending order (per count)


print ("The number of emails received = ", len(email_body)) #This tells there were total of 14 emails received in 
# Beck's Australia folder
```

```python
#Importing Ken Lays emails into 3 dictionaries: To_Email, From_Email, and Email_Body

from email.parser import Parser
rootdir3 = r'/home/EDSE-suaniarh-workspace/Enron_dataset/maildir/lay-k/all_documents' 


to_email_dict = {}
from_email_dict = {}
email_body = {}


def email_analyse2(inputfile, to_email_dict, from_email_dict, email_body):
    with open(inputfile, "r") as f:
        data = f.read()

    email = Parser().parsestr(data)
       
    if email['to']:
        email_to = email['to']
        email_to = email_to.replace("\n", "")
        email_to = email_to.replace("\t", "")
        email_to = email_to.replace(" ", "")

        email_to = email_to.split(",")

        
            
        for mail in email_to:
            if mail in to_email_dict.keys():
                to_email_dict[mail] += 1 ##If the email address already exists, it will add 1 to it......
            else:
                to_email_dict[mail] = 1 ##.......else it will record that entry as 1
                
        
    if email['from'] in from_email_dict.keys():
        from_email_dict[email['from']]+= 1 
        
    else:
        from_email_dict[email['from']]= 1            
            
start = time.time()

for directory, subdirectory, filenames in os.walk(rootdir2):
    for filename in filenames:
        email_analyse2(os.path.join(directory, filename), to_email_dict, from_email_dict, email_body)
        
print(from_email_dict, "\n", "\n")

print(to_email_dict)
```

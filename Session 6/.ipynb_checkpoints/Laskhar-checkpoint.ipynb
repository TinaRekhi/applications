```python
#importing libraries and modules to assist with the code

#importing pandas to place data in the table format
import pandas as pd 

#importing bokeh and its modules to build interactive graphs
from bokeh.plotting import figure, output_notebook, show 

#This allows inline plotting for Jupyter Notebook
output_notebook() 

from bokeh.models import Range1d

#Importing math to use mathematical functions
import math 
```

```python
# Datasheet was downbloaded form GTD and being read in the form of a local .csv file
GTD = pd.read_excel(r"/home/EDSE-suaniarh-workspace/GTD_db/globalterrorismdb_0919dist.xlsx") 

# If there is no number between the parethesis, it pulls the first five rows by default, which is what we want here.
GTD.head()
```

```python
# Reducing the dataframe down to just India and storing the information in an object called country
country = GTD[GTD["country_txt"]=="India"] 

# Changing the default setting of 1st 5 rows to 1st 15 rows
country.head(15) 
```

```python
# Creating a dicitionary called attacks_by_group to identify the various terrorist groups active in India along and 
# want to calculate teh total number of attacks by each one of them.

attacks_by_group = {} 

#This loop goes through each row and counts the number of entries by group
for index, row in country.iterrows(): 
    if row["gname"]  in attacks_by_group.keys():
        attacks_by_group[row["gname"]] += 1 #if group is in the dictionary add 1 attack
    else:
        attacks_by_group[row["gname"]] = 1 #add group name to dictionary if not in dictionary

attacks_by_group 
```

```python
# Further reducing the dataframe down to focus only one one Terrorist Group active in India - Lashkar e Taiba aka LeT
# Storing the information in an object called LashkarIndia

lashkarIndia = country[country["gname"]=="Lashkar-e-Taiba (LeT)"]
lashkarIndia.head(15)
```

```python
# Focusing on Lashkar-e-Taiba to see what other countries have they attacked on besides India. 
# Therefore reduced the dataframe down to just that terrorist group from the main GTD database.
# Stored this information in an object called terrorist.

terrorist = GTD[GTD["gname"]=="Lashkar-e-Taiba (LeT)"] 
terrorist.head(10) # limiting it to 10 rows.
```

```python
#Now I'd like to see what other countries has this group attacked within. For this
# I'll create a dictionary with keys as the Country names and values as the number of attacks

attacks_by_lashkar = {} 

for index, row in terrorist.iterrows(): 
    if row["country_txt"]  in attacks_by_lashkar.keys():
        attacks_by_lashkar[row["country_txt"]] += 1 #if that country is in the dictionary then add 1 attack
    else:
        attacks_by_lashkar[row["country_txt"]] = 1 #add the country name if not already in the dictionary

attacks_by_lashkar 
```

```python
# Plotting attacks by LeT on a bar chart using data captured above. 
countries = list(attacks_by_lashkar.keys())
attacks = list(attacks_by_lashkar.values())

#Before plotting, would  like to sort the data between the three countries (Afghanistan, Pakistan, and India). 

sorted_countries = sorted(countries, key=lambda x: attacks[countries.index(x)])
sorted_attacks = sorted(attacks)

tsort_attacks=(sorted_attacks[:])
tsort_countries=(sorted_countries[:])
```

```python
# Used the bokeh library to plot an interactive graph using the data above to make a figure.

p = figure(x_range = tsort_countries, plot_width=400, plot_height=400)

p.vbar(x=tsort_countries, width=0.20, bottom=0,
       top= tsort_attacks, color="purple")

p.xaxis.major_label_orientation = math.pi/2  #Rotates the labels on the X axis

show(p)
```

```python
# Now focusing back on LeT and it's attacks within India (their favorite target country).
#Creating a dictionary to calculate the number of attacks by LeT sorted by year in India. 
attacks_by_year = {} 

#Created a for loop that goes through each row to count the number of entries against LeT by year.

for index, row in lashkarIndia.iterrows(): 
    if row["iyear"]  in attacks_by_year.keys():
        attacks_by_year[row["iyear"]] += 1 #if group is in the dictionary add 1 attack
    else:
        attacks_by_year[row["iyear"]] = 1 #add group name to dictionary if not in dictionary

attacks_by_year
```

```python
# Plotting details from above into a graph. 

eventDate = list(attacks_by_year.keys())
eventAttacks = list(attacks_by_year.values())

import matplotlib.pyplot as plt

plt.plot(eventDate, eventAttacks, linewidth=5.0)
plt.title('Total Attacks by Lashkar e Taiba (LeT) in India', color = "brown", fontweight="bold")
plt.xlabel('Year', color="brown", fontweight="bold")
plt.ylabel('Number of Attacks', color = "brown", fontweight="bold")
plt.show()
```

```python
#Plotting on a map using bokeh and the Stamen Terrain Map. Importing transformer to convert longitude and latitudes.
from bokeh.tile_providers import get_provider, Vendors
from pyproj import Transformer
tile_provider = get_provider('STAMEN_TERRAIN')
```

```python
# But first we'll need to oganize the data for plotting.
# Collecting the following from the object India to plot: (1) Lat/Long of attacks (2) event ID and (3) year.
country_map = country[["latitude", 'longitude', 'eventid', 'iyear']]

country_map.head()
```

```python
#To convert from lat/long to web mercator
transformer = Transformer.from_crs('epsg:4326','epsg:3857')
```

```python
map_dict = {} # empty dictionary to track events by lat long
nan_count = {} # empty dictionary to track events without a lat/long entry to see how many events are we not 
#going to be able to track as a result.


# Iterating through tables and associating each event with lat/long
for idx, row in country_map.iterrows():    
    if row['eventid'] in map_dict.keys(): 
        if math.isnan(row["latitude"]):
            #This counts no data
            if row['eventid'] in nan_count.keys(): 
                nan_count[row['eventid']] += 1 
            else: 
                nan_count[row['eventid']] = 1
        else: 
            #To convert the lat/long to a mercator projection 
            point = transformer.transform(row["latitude"],row["longitude"])
            map_dict[row['eventid']].append([point[0],point[1]]) 
            
    
    else:  
        if math.isnan(row["latitude"]):
          nan_count[row['eventid']] = 1
        else: 
            point = transformer.transform(row["latitude"],row["longitude"])
            map_dict[row['eventid']] =[[point[0],point[1]]]
        
#This tells how many attacks we are losing
nan_count  
len(nan_count)
```

```python
#Entering Lat/Long coordinates for India to plot attacks on the map.

pts = [(3.25,58.67), (42, 97.7)]
bbox = []
for pt in transformer.itransform(pts): 
    bbox.append(pt)
```

```python
LeT_x = []
LeT_y = []
for k,v in map_dict.items(): 
    for pt in v:
        LeT_x.append(pt[0])
        LeT_y.append(pt[1])
```

```python
#To plot the bounding box
p = figure(x_range=(bbox[0][0], bbox[1][0]),y_range=(bbox[0][1], bbox[1][1]),x_axis_type="mercator", y_axis_type="mercator")

# Adding the map from the Bokeh map vendor - Stamen_Terrain
p.add_tile(tile_provider)

# To place a circle for each converted lat/long attack 
p.circle(x = LeT_x, y = LeT_y, color= "firebrick")

#To show the plot
show(p)
```

```python
# I created a subset for every 5 years: 1995-2000, 2000-2005, 2005-2010, 2010-2015, 2015-2020.

a_1995 = country_map[(country_map['iyear']>=1995) & (country_map['iyear']<2000)]


b_2000 = country_map[(country_map['iyear']>=2000) & (country_map['iyear']<2005)]

c_2005 = country_map[(country_map['iyear']>=2005) & (country_map['iyear']<2010)]

d_2010 = country_map[(country_map['iyear']>=2010) & (country_map['iyear']<2015)]

e_2015 = country_map[(country_map['iyear']>=2015) & (country_map['iyear']<2020)]
```

```python
# And this is how I plot it using a different function called country_map_convert, which we will use in the next cell
# to plot for each of the 5 subsets above.

def country_map_convert(country_map):
    
    map_dict = {} # empty dictionary to track events by lat long
    nan_count = {} # empty dictionary to track events without a lat/long entry to see how many events are we not 
    #going to be able to track as a result.


    # Iterating through tables and associating each event with lat/long
    for idx, row in country_map.iterrows():    
        if row['eventid'] in map_dict.keys(): 
            if math.isnan(row["latitude"]):
                #This counts no data
                if row['eventid'] in nan_count.keys(): 
                    nan_count[row['eventid']] += 1 
                else: 
                    nan_count[row['eventid']] = 1
            else: 
                #To convert the lat/long to a mercator projection 
                point = transformer.transform(row["latitude"],row["longitude"])
                map_dict[row['eventid']].append([point[0],point[1]]) 


        else:  
            if math.isnan(row["latitude"]):
              nan_count[row['eventid']] = 1
            else: 
                point = transformer.transform(row["latitude"],row["longitude"])
                map_dict[row['eventid']] =[[point[0],point[1]]]

    #This tells how many attacks we are losing
    
    print(len(nan_count))
    
    LeT_x = []
    LeT_y = []
    for k,v in map_dict.items(): 
        for pt in v:
            LeT_x.append(pt[0])
            LeT_y.append(pt[1])
    
    return LeT_x, LeT_y
    
```

```python
# This is creating x and Y coordinates for each of the 5 subsets (time series) identified in the cells above.

x_1995, y_1995 = country_map_convert(a_1995)
x_2000, y_2000 = country_map_convert(b_2000)
x_2005, y_2005 = country_map_convert(c_2005)
x_2010, y_2010 = country_map_convert(d_2010)
x_2015, y_2015 = country_map_convert(e_2015)
```

```python
#To plot the bounding box
p = figure(x_range=(bbox[0][0], bbox[1][0]),y_range=(bbox[0][1], bbox[1][1]),x_axis_type="mercator", y_axis_type="mercator")

# Adding the map from the Bokeh map vendor - Stamen_Terrain
p.add_tile(tile_provider)

# To place a circle for each converted lat/long attack 
p.circle(x = x_1995, y = y_1995, color= "purple")
p.circle(x = x_2000, y = y_2000, color= "blue")
p.circle(x = x_2005, y = y_2005, color= "red")
p.circle(x = x_2010, y = y_2010, color= "black")
p.circle(x = x_2015, y = y_2015, color= "yellow")


#To show the plot
show(p)
```
